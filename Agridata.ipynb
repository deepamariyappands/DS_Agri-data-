```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Aggregate rice production by state
rice_production_by_state = df.groupby("state_name")["rice_production_(1000_tons)"].sum().reset_index()

# Sort and select top 7 states
top_7_rice_states = rice_production_by_state.sort_values(by="rice_production_(1000_tons)", ascending=False).head(7)

# Plot bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x="state_name", y="rice_production_(1000_tons)", data=top_7_rice_states, palette="viridis")

plt.xlabel("State", fontsize=12)
plt.ylabel("Rice Production (1000 Tons)", fontsize=12)
plt.title("Top 7 Rice Producing States in India", fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

```python
#Top 5 Wheat Producing States Data(Bar_chart)and its percentage(%)(Pie_chart)

# Aggregate wheat production by state
wheat_states = df.groupby("state_name")["wheat_production_(1000_tons)"].sum().reset_index()

# Select top 5 wheat-producing states
top_5_wheat_states = wheat_states.nlargest(5, "wheat_production_(1000_tons)")

# Bar Chart
plt.figure(figsize=(10, 5))
sns.barplot(x="state_name", y="wheat_production_(1000_tons)", data=top_5_wheat_states, palette="Blues_r")
plt.xlabel("State")
plt.ylabel("Wheat Production (1000 Tons)")
plt.title("Top 5 Wheat Producing States in India")
plt.xticks(rotation=45)
plt.show()

# Pie Chart
plt.figure(figsize=(7, 7))
plt.pie(top_5_wheat_states["wheat_production_(1000_tons)"], labels=top_5_wheat_states["state_name"], autopct="%1.1f%%", colors=sns.color_palette("Blues_r", 5))
plt.title("Wheat Production Percentage (Top 5 States)")
plt.show()
```

```python
# Get the top 5 oilseed-producing states

# Assuming df is your DataFrame
top_oilseeds = df.groupby("state_name")["oilseeds_production_(1000_tons)"].sum().reset_index()
top_oilseeds = top_oilseeds.sort_values(by="oilseeds_production_(1000_tons)", ascending=False).head(5)

# Lollipop Chart
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

plt.stem(top_oilseeds["state_name"], top_oilseeds["oilseeds_production_(1000_tons)"], linefmt='grey', markerfmt='ro', basefmt=" ")

# Labels and Title
plt.xlabel("State")
plt.ylabel("Oilseed Production (1000 tons)")
plt.title("Top 5 Oilseed Producing States in India")
plt.xticks(rotation=45)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Aggregate sunflower production by state
sunflower_production = df.groupby("state_name")["sunflower_production_(1000_tons)"].sum().reset_index()

# Get top 7 states by sunflower production
top_7_sunflower_states = sunflower_production.sort_values(by="sunflower_production_(1000_tons)", ascending=False).head(7)

# Plot area chart
plt.figure(figsize=(12, 6))
plt.fill_between(top_7_sunflower_states["state_name"], top_7_sunflower_states["sunflower_production_(1000_tons)"], color="gold", alpha=0.6)

plt.plot(top_7_sunflower_states["state_name"], top_7_sunflower_states["sunflower_production_(1000_tons)"], marker="o", color="brown", linestyle="dashed")

plt.xlabel("State")
plt.ylabel("Sunflower Production (1000 tons)")
plt.title("Top 7 Sunflower Producing States in India")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.show()
```

```python
#India's SUGARCANE PRODUCTION From Last 50 Years(Line_plot)

# Select relevant columns
sugarcane_df = df[['year', 'sugarcane_production_(1000_tons)']]

# Aggregate production by year (if required)
sugarcane_production = sugarcane_df.groupby('year').sum().reset_index()

# Plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=sugarcane_production, x='year', y='sugarcane_production_(1000_tons)', marker='o', color='green')

plt.xlabel('Year')
plt.ylabel('Sugarcane Production (1000 tons)')
plt.title("India's Sugarcane Production Over the Last 50 Years")
plt.grid(True)
plt.show()
```

```python
#Rice Production Vs Wheat Production (Last 50y)


# Select relevant columns
df_filtered = df[['year', 'rice_production_(1000_tons)', 'wheat_production_(1000_tons)']]

# Aggregate total production over the last 50 years
df_agg = df_filtered.groupby('year').sum()

# Sum total rice and wheat production
total_rice = df_agg['rice_production_(1000_tons)'].sum()
total_wheat = df_agg['wheat_production_(1000_tons)'].sum()

# Pie chart
labels = ['Rice Production', 'Wheat Production']
sizes = [total_rice, total_wheat]
colors = ['skyblue', 'lightcoral']  # Skyblue for Rice, Rose (lightcoral) for Wheat

plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140, wedgeprops={'edgecolor': 'black'})
plt.title("Rice vs Wheat Production (Last 50 Years)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# Define crops and their respective columns
crops = {
    "Rice": ("rice_area_(1000_ha)", "rice_production_(1000_tons)"),
    "Wheat": ("wheat_area_(1000_ha)", "wheat_production_(1000_tons)"),
    "Maize": ("maize_area_(1000_ha)", "maize_production_(1000_tons)")
}

# Set plot style
sns.set_style("whitegrid")
plt.figure(figsize=(15, 5))

# Create scatter plots with regression lines
for i, (crop, (area_col, prod_col)) in enumerate(crops.items(), 1):
    plt.subplot(1, 3, i)
    sns.regplot(x=df[area_col], y=df[prod_col], scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
    plt.title(f"{crop} - Area vs Production")
    plt.xlabel("Area Cultivated (1000 ha)")
    plt.ylabel("Production (1000 tons)")

plt.tight_layout()
plt.show()
```

```python
#Rice vs. Wheat Yield Across States 

# Selecting relevant columns
yield_data = df[['state_name', 'rice_yield_(kg_per_ha)', 'wheat_yield_(kg_per_ha)']]

# Aggregating data at the state level
yield_data = yield_data.groupby('state_name', as_index=False).mean()

# Reshaping data for visualization
yield_melted = yield_data.melt(id_vars='state_name', 
                               var_name='Crop', 
                               value_name='Yield (kg/ha)')

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x='state_name', y='Yield (kg/ha)', hue='Crop', data=yield_melted)
plt.xticks(rotation=90)
plt.xlabel('State')
plt.ylabel('Yield (kg/ha)')
plt.title('Rice vs. Wheat Yield Across States')
plt.legend(title='Crop', labels=['Rice', 'Wheat'])
plt.show()
```
